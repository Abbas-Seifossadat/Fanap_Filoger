For missing value handling in the preprocessing step, we can fill null values of <strong>dep_delay</strong> from one of the famous methods such as mean. In previous insights, we mentioned that the <strong>dep_delay</strong> is a good representation of the <strong>sched_dep_time</strong> and <strong>dep_time</strong>. So, for missing values handling in <strong>dep_time</strong>, we can fill them with the sum of <strong>dep_delay</strong> and <strong>sched_dep_time</strong> values.</li>

```python
# using astype(str) for converting the type of flight
data['flight'] = data['flight'].astype(str)
```

<div style="background-color:#fdf6e3; padding:3px; border-style: solid; font-family:'Comic Sans MS', cursive; color:#000080; font-size:100%; text-align:center; border-radius:20px 60px;">
    <strong>üí≠Insightsüí≠</strong>:
    <ul style="text-align: left;">
        <li>According to <code>data.info()</code>, we see that <strong>dep_time</strong>, <strong>sched_dep_time</strong>, <strong>arr_time</strong>, <strong>and sched_arr_time</strong> columns have <em>float</em> and <em>int</em> Dtypes. These columns represent the time. Thus, if we consider them as <em>int</em> or <em>float-based</em>, it seems non-logical. Therefore, by converting them to a time-based data type, we'll be able to utilize various time-related functionalities more effectively.</li>
    </ul>
</div>

```python
columns_to_convert = ['arr_time', 'dep_time', 'sched_arr_time', 'sched_dep_time']

'''
This code presents these conditions:
1. Convert columns_to_convert (float Dtype) to integer
2. Convert columns_to_convert to string
3. the hour component should be between 00 and 23, So replace('2400', '0000')
4. Add leading zeros to ensure consistent length (e.g., 517 -> '0517')
5. Convert to time-based data type
'''

for column in columns_to_convert:
    data[f'{column}_timebased'] = data[column].apply(lambda x:
                                                      pd.NaT if pd.isnull(x)
                                                      else pd.to_datetime(str(int(x)).replace('2400', '0000').zfill(4),
                                                      format='%H%M', errors='coerce').strftime('%H:%M'))

# checking the info after changing the Dtypes
columns = ['arr_time', 'dep_time', 'sched_arr_time', 'sched_dep_time',
          'arr_time_timebased', 'dep_time_timebased', 'sched_arr_time_timebased', 'sched_dep_time_timebased']
data[columns].info()
```

<div style="background-color:#fdf6e3; padding:3px; border-style: solid; font-family:'Comic Sans MS', cursive; color:#000080; font-size:100%; text-align:center; border-radius:20px 60px;">
    <strong>üí≠Insightsüí≠</strong>
    <ul style="text-align: left;">
        <li>According to the above discussion, I decided to drop these columns:
          <ol>
              <li><strong>arr_time</strong>         ‚ñ∂ Since I create a new column (arr_time_timebased)</li>
              <li><strong>dep_time</strong>         ‚ñ∂ Since I create a new column (dep_time_timebased)</li>
              <li><strong>sched_arr_time</strong>   ‚ñ∂ Since I create a new column (sched_arr_time_timebased)</li>
              <li><strong>sched_dep_time</strong>   ‚ñ∂ Since I create a new column (sched_dep_time_timebased)</li>
              <li><strong>time_hour</strong>        ‚ñ∂ Since It is a redundant feature</li>
          </ol>
        </li>
    </ul>
</div>


```python
data.drop(['arr_time', 'dep_time', 'sched_arr_time', 'sched_dep_time', 'time_hour'], axis=1, inplace=True)
```

<p style="font-family:Comic Sans MS', cursive; font-size:120%; color:#2AAA8A;">For univariate analysis, the Trimmed mean is a good choice for better understanding the data.
</p>

```python
num_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Calculate mean and trimmed mean for each numerical column
mean_vs_trimmed_mean = pd.DataFrame({'mean': data[num_columns].mean(),
                                     'trimmed_mean': data[num_columns].apply(lambda x: trim_mean(x, proportiontocut=0.1))})

# Calculate kurtosis and skewness for each numeric column
kurtosis_skewness = pd.DataFrame({'kurtosis': data[num_columns].apply(kurtosis),
                                  'skewness': data[num_columns].apply(skew)})

print("Mean vs Trimmed Mean:")
print(mean_vs_trimmed_mean)

print("\nKurtosis and Skewness:")
print(kurtosis_skewness)
```

<div style="background-color:#fdf6e3; padding:3px; border-style: solid; font-family:'Comic Sans MS', cursive; color:#000080; font-size:100%; text-align:center; border-radius:20px 60px;">

    <ul style="text-align: left;">

        <li>In addition, As we expected the <strong>month</strong>, <strong>day</strong>, <strong>Hour</strong>, and <strong>minute</strong> features have a <strong>negative kurtosis</strong> (Platykurtic). So, the flights were distributed all over the year and at different times (These features approximately operate like uniform distribution).</li>

    The following snippet code determines the percentage of outliers in <strong>dep_delay</strong>, <strong>arr_delay</strong>, <strong>air_time</strong>, and <strong>distance</strong> features.
</div>

```python
# outlier data
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
# outlier data
columns = ['dep_delay', 'arr_delay', 'air_time', 'distance']
outlier_detection = data[columns]
for k, v in outlier_detection.items():
        outliers = find_outliers_IQR(v)
        perc = len(find_outliers_IQR(v)) * 100.0 / np.shape(outlier_detection)[0]
        print(f'Column :{k}')
        print(f"Number of Outliers = {len(outliers)}","||",f"Max Outlier Value = {outliers.max()}",
              "||", f"Min Outlier Value = {outliers.min()}", "||", f"Percentage of Outliers = {perc:.2f}%")
        # print(f"Max Outlier Value = {outliers.max()}")
        # print(f"Min Outlier Value = {outliers.min()}")
        # print(f"Percentage of Outliers = {perc:.2f}%")
        print("\n")
```

<div style="font-family:Comic Sans MS', cursive; color:#2AAA8A; font-size:120%">
    The above chart shows the destinations of flights in this dataset. The following snippet code determines which destination has more delay:
</div>

```python
# Group by 'dest' and calculate the trimmed mean for 'arr_delay'
dest_delay_means = data.groupby('dest').agg(trimmed_mean_delay=('arr_delay', lambda x: trim_mean(x, proportiontocut=0.1)))

# Reset the index to access the 'dest' column
dest_delay_means = dest_delay_means.reset_index()

# Determine the top 5 destinations with the most delay
top_delay_destinations = dest_delay_means.nlargest(5, 'trimmed_mean_delay')

# Determine the top 5 destinations with the lowest delay
top_least_delay_destinations = dest_delay_means.nsmallest(5, 'trimmed_mean_delay')

print("Top 5 destinations with the most delay:")
print(top_delay_destinations)

print("\nTop 5 destinations with the lowest delay:")
print(top_least_delay_destinations)
```
